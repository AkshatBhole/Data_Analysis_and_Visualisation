```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(1,10,15)
x
```

```python
y=x**7
y
```

```python
plt.plot(x,y)
plt.title("Chart")
plt.xlabel("Chacha")
plt.ylabel("Chachi")
```

```python
# SUBPLOT
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.subplot(2,2,4)
```

```python
# Object oriented Way
```

```python
fig=plt.figure()
axis1=fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)

axis2=fig.add_axes([0.1,0.1,0.3,0.3])
axis2.plot(x,y)
```

```python
plt.scatter(x,y)
```
