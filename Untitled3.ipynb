# distribution plots are used in quantitative data like total_bill tip

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=sns.load_dataset("tips")
df
```

```python
df['size'].unique()
```

# DISPLOT

```python
sns.displot(df['total_bill'],bins = 1)
```

```python
sns.displot(df['total_bill'],bins = 100)
```

```python
sns.displot(df['total_bill'],bins = 100,kde=True)     #kernal distribution means we get skewed 
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 100,kde=True)     #kernal distribution means we get skewed 
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 100,kde=True)     #kernal distribution means we get skewed 

```

# JOINT-PLOT

```python
# do data ko compare kara chahta hu histplot me alag lah aate the subplot se 
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

# PAIR_PLOT                   ####pair kar deta multiple chizo ko

```python
sns.pairplot(df,hue='size',palette='rainbow')            #numerical columns ka comparizon
```

#RUG_PLOT

```python
sns.rugplot(df)
```
